```python
# 1.create user defined to understand the column
def info_of_cat(col):
    print(f"unique values in {col} are:{df[col].nunique()}") #unique willl return the unique value of each column
    print(f"Mode values in {col} is:{df[col].mode()[0]}") #it will return mode pf a column
    print(f"number of missing value in {col} are :{df[col].isnull().sum()}") #it will return the count of missing values
    
```

```python
# 2.analayse the required column:Attrition_Flag
# 2.1:use functions to get the information about Attrition_Flag
info_of_cat('Attrition_Flag')
```

```python
# 2.2 Visualize it 
fig,ax=plt.subplots(1,2,figsize=(15,7))
# Title
ax[0].set_title("Count_Of_attrition_Flag")
ax[1].set_title('pie-chart')
percentage=df["Attrition_Flag"].value_counts()

lab=list(df["Attrition_Flag"].value_counts().index)

sns.countplot(x=df["Attrition_Flag"],ax=ax[0])
plt.pie(percentage,labels=lab,autopct="%.2f%%")
plt.show()
```

```python
# ax.patches
# it returns the rectangle object from matplotlib 
# annotate is used to represents the labels over the objects here objects are the bars
# vertival allignment =perecentage ove the bar head
# horizontal allignment=percentage over the center
# get_x:
# it returns the left coordinate of the rectangele
# get_width:
# it returns the width of the reactangle

```

```python
fig, ax  = plt.subplots(1, figsize = (15,7))
ax = sns.countplot(x = df["Dependent_count"])

for bar in ax.patches: 
    percentage = f"{round(bar.get_height()/len(df), 2)}%"
    x = bar.get_x() + bar.get_width() /2
    y = bar.get_height()
    ax.annotate(percentage,(x, y), va ="bottom" , ha ="center")
plt.show()
```
