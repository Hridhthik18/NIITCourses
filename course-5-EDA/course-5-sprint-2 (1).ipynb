```python
import matplotlib.pyplot as plt
import matplotlib as mat
import pandas as pd
import numpy as np
from tabulate import tabulate
import seaborn as sns
```

```python
df=pd.read_csv('DS1_C5_S1_EmployeeAttrition_Data_Practice.csv')
df
```

```python
cat_col=df.select_dtypes(include='object')
num_col=df.select_dtypes(exclude='object')
list_num=list(num_col)
list_cat=list(cat_col)
print(tabulate({'categorical':list_cat,'Numerical':list_num},headers=['Categorical','Numerical']))
```

```python
# level 0 univariate analysis:
def cat_display(col):
    print(f'{col} has {df[col].nunique()} has unique numbers')
    print(f'{col} has {df[col].mode()[0]} has mode values')
    print(f'{col} has {df[col].isnull().sum()} has null values')
    if df[col].isnull().sum()>0:
        df[col]=df[col].fillna(df[col].mode()[0])
        print('missing values replaced')
    
    print('---------------------------------------------------------')

```

```python
for i in list_cat:
    cat_display(i)
    
```

```python
def num_display(col):
    print(f'{col} has {df[col].nunique()} has unique numbers')
    print(f'{col} has {df[col].mean()} has mean values')
    print(f'{col} has {df[col].isnull().sum()} has null values')
    if df[col].isnull().sum()>0:
        df[col]=df[col].fillna(df[col].mean())
        print('missing values replaced')
    
    print('---------------------------------------------------------')
```

```python
for j in list_num:
    num_display(j)
    
```
