# LEVEL 1: UNIVARIATE ANALYSIS:  CATEGORICAL

```python
#1. Create user defined function to understand the column

def info_of_cat(col):
    print(f"unique values in {col} are : {df[col].nunique()}")   # unique will return the unique value of each column
    print(f"Mode value in {col} is: {df[col].mode()[0]}")        # It will return Mode pf a column
    print(f"Number of missing value in {col} are : {df[col].isnull().sum()}")   #It will return the count of missing values
```

```python
#2. Anlayse the required columns : Attrition_Flag
#2.1 : Use function to get the information about attrition flag
info_of_cat("Attrition_Flag")
```

```python
#2.2. Visualise it
fig, ax  = plt.subplots(1,2, figsize = (15,7))
ax[0].set_title("Count of attrition Flag")    #set title
percentage = df["Attrition_Flag"].value_counts()
lab = list(df["Attrition_Flag"].value_counts().index)

sns.countplot(x = df["Attrition_Flag"], ax = ax[0])
plt.pie(percentage,  labels = lab, autopct = "%.2f%%")

plt.show()
```

```python
fig, ax  = plt.subplots(1, figsize = (15,7))
ax = sns.countplot(x = df["Dependent_count"])
plt.show()

#It is not possible than dependen count will be greater than 5. So we will deal with all with the average value
```

```python
#So calculate average value
mean = df.Dependent_count.mean()
print(mean)
```

```python
 # Below are the data values available in df greater than 5
df[df["Dependent_count"] >5]  
```

```python
#access all rows greater than 5 and access location also
x = df[df["Dependent_count"] >5].index
print(x)
```

```python
#replace all data values with mean which are greater than 5 dependent count
for index in x:
    df.loc[index, "Dependent_count" ] = mean
```

```python
#Validate again. Check whether the mean values has replaced in df
df[df["Dependent_count"] >5]

#now, all dependent count greater than 5 has replaced with the average value: 2.7
```

```python
#Again visualise and interpret
fig, ax  = plt.subplots(1, figsize = (15,7))
ax = sns.countplot(x = df["Dependent_count"])

for bar in ax.patches: 
    percentage = f"{round(bar.get_height()/len(df), 2)}%"
    x = bar.get_x() + bar.get_width() /2
    y = bar.get_height()
    ax.annotate(percentage,(x, y), va ="bottom" , ha ="center")
plt.show()
```
