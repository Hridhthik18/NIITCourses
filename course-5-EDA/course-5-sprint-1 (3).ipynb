```python
# task2
from tabulate import tabulate

# Lists to store column names by type
mylist1 = []
mylist2 = []

# Function to categorize columns by data type
def type_info():
    # Selecting categorical columns
    cat_col = df.select_dtypes(include='object').columns
    mylist1.extend(cat_col)  # Append column names, not the DataFrame

    # Selecting numerical columns
    num_col = df.select_dtypes(exclude='object').columns
    mylist2.extend(num_col)  # Append column names, not the DataFrame

# Call the function to populate mylist1 and mylist2
type_info()

# Display the lists in a tabulated format
print(tabulate({'Categorical': mylist1, 'Numerical': mylist2}, headers=['Categorical', 'Numerical']))

```
