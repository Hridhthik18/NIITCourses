```python
# task1
# Display basic information about the dataset, which includes column names, number of rows and columns, and data type of the columns. 
import pandas as pd

df = pd.read_csv('DS1_C5_S3_Sales_Data_Practice.csv')
print(df.info())
print(df.shape)
print(df.dtypes)    
print(df.columns)
print(df.head())
print(df.tail())
# Also, display the count of missing values for each column.
print(df.isnull().sum())
```

```python
# task2
# Separate the columns into categorical and numerical data types and display them.
print(df.select_dtypes(include=['object']).columns)
print(df.select_dtypes(include=['int64', 'float64']).columns)
# Display the summary statistics of the numerical columns.
print(df.describe())
# Display the summary statistics of categorical columns.
print(df.describe(include=['object']))
```

```python
# Conduct descriptive analysis of numerical columns 'PRICE EACH', 'SALES', and 'MSRP' to determine the unique values present in each column, their respective modes, and other relevant statistics.
print(df['PRICEEACH'].unique())
print(df['PRICEEACH'].mode())
print(df['PRICEEACH'].value_counts())
print(df['PRICEEACH'].describe())
print(df['SALES'].unique())
print(df['SALES'].mode())
print(df['SALES'].value_counts())
print(df['SALES'].describe())
print(df['MSRP'].unique())
print(df['MSRP'].mode())
print(df['MSRP'].value_counts())
print(df['MSRP'].describe())
```

```python
# Perform descriptive analysis of categorical columns 'PRODUCTLINE', 'CUSTOMERNAME', and 'COUNTRY' to identify the unique values present in each column, their respective modes, and other relevant statistics.
print(df['PRODUCTLINE'].unique())
print(df['PRODUCTLINE'].mode())
print(df['PRODUCTLINE'].value_counts())
print(df['PRODUCTLINE'].describe())
print(df['CUSTOMERNAME'].unique())
print(df['CUSTOMERNAME'].mode())
print(df['CUSTOMERNAME'].value_counts())
print(df['CUSTOMERNAME'].describe())
print(df['COUNTRY'].unique())
print(df['COUNTRY'].mode())
print(df['COUNTRY'].value_counts())
print(df['COUNTRY'].describe())
```

```python
# task3
# Perform descriptive analysis of the numerical columns 'PRICE EACH', 'SALES', and 'MSRP' by using the following: Bar chart, Pie chart
import matplotlib.pyplot as plt
df['PRICEEACH'].value_counts().plot(kind='bar')
plt.show()
df['PRICEEACH'].value_counts().plot(kind='pie')    
plt.show()
df['SALES'].value_counts().plot(kind='bar')
plt.show()
df['SALES'].value_counts().plot(kind='pie')
plt.show()
df['MSRP'].value_counts().plot(kind='bar')
plt.show()
df['MSRP'].value_counts().plot(kind='pie')
plt.show()
```

```python
# Perform descriptive analysis of the numerical columns 'PRICE EACH', 'SALES', and 'MSRP' by using the following: Measures of central tendency (mean, median)
print(df['PRICEEACH'].mean())
print(df['PRICEEACH'].median())
print(df['SALES'].mean())
print(df['SALES'].median())
print(df['MSRP'].mean())
print(df['MSRP'].median())
```

```python
# Perform descriptive analysis of the numerical columns 'PRICE EACH', 'SALES', and 'MSRP' by using the following:Standard deviation of the column
print(df['PRICEEACH'].std())
print(df['SALES'].std())
print(df['MSRP'].std())
```

```python
# Perform descriptive analysis of the numerical columns 'PRICE EACH', 'SALES', and 'MSRP' by using the following:Skewness of the column
print(df['PRICEEACH'].skew())
print(df['SALES'].skew())
print(df['MSRP'].skew())
```
