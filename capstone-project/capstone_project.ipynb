```python
# appearances_df and games_df
merge_1=pd.merge(appearances_df,games_df,on='game_id',how='inner')
merge_1
```

```python
# merged1 and game_events_df
merge_2=pd.merge(merge_1,game_events_df,on=['game_id','player_id'],how='inner')
merge_2
```

```python
# rename for the player_id_x  to player_id because we are getting null values
merge_2=merge_2.rename(columns={'player_id_x':'player_id'})
merge_2
```

```python
# merge_2 and game_lineups_df
merge_3=pd.merge(merge_2,game_lineups_df,on='player_id',how='inner')
merge_3
```

```python
# merge_3 and players_df
merge_4=pd.merge(merge_3,players_df,on='player_id',how='inner')
merge_4
```

```python
df=merge_4.drop(['date_x','date_y','competition_id_y','player_name_x','player_name_y'],axis=1)
df
```

```python
# missing value treatment
```

```python
df.isnull().sum()
```

```python
df=df.drop(['home_club_formation','away_club_formation','player_assist_id'],axis=1)
df
```

```python
numerical_col=df.select_dtypes(exclude='object')
categorical_col=df.select_dtypes(include='object')


numerical_lst=list(numerical_col)
categorical_lst=list(categorical_col)

print(tabulate({'Numerical':numerical_lst,'categorical':categorical_lst},headers=['Numerical','categorical']))
```

```python
# numerical columns are replaced by mean values
for i in numerical_col:
    df[i]=df[i].fillna(df[i].mean())
        
```

```python
# categorical_columns are replaced by mode value
for j in categorical_col:
    df[j]=df[j].fillna(df[j].mode()[0])
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)
plt.gca().set_facecolor('lightgrey')
sns.boxplot(data=df,color='gold')
```

```python
# takimg numerical columns from original dataframe
num=df.select_dtypes(exclude='object')
num
```

```python
data=[]
for i in num:
    uo=[]
    lo=[]
    ser=num[i]
    Min=ser.min()
    Max=ser.max()
    Range=Max-Min
    q1=ser.quantile(0.25)
    q3=ser.quantile(0.75)
    iqr=q3-q1
    uf=q3+1.5*iqr
    lf=q1-1.5*iqr
    for z in ser:
        if z>uf:
            uo.append(z)
        if z<lf:
            lo.append(z)
    data.append([i,Min,Max,Range,q1,q3,iqr,uf,lf,[uo,lo]])
out_df=pd.DataFrame(data,columns=['column','min','max','range','q1','q3','iqr','uf','lf','outliers'])
out_df
```

```python
# outlier treatment for number column
def treat(col):
    ser=df[col]
    q1=ser.quantile(0.25)
    q3=ser.quantile(0.75)
    iqr=q3-q1
    uf=q3+1.5*iqr
    lf=q1-1.5*iqr
    for z in ser:
        if z>uf or z<lf:
            df[col]=df[col].replace(z,df[col].mean())
            
            
```

```python
treated=treat('number')
treated
```
