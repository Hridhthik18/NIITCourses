```python
def calculate_residuals(model,feature,labels):
    predictions=model.predict(feature)
    df_result=pd.DataFrame({'actual':labels,'predicted':predictions})
    df_result['residuals']=abs(df_result['actual'])-abs(df_result['predicted'])
    return df_result
def linear_assumption(model,feature,labels):
    df_result=calculate_residuals(model,feature,labels)
    sns.lmplot(x='actual',y='predicted',data=df_result,fit_reg=False)
    coords=np.arange(df_result.min().min(),df_result.max().max())
    plt.plot(coords,coords,color='red',ls='--')
    plt.show()

# calculate_residuals computes actual vs. predicted values and their residuals (errors).
# linear_assumption visualizes predictions against actual values to check linearity.
# The dashed red line represents a perfect fit; deviations indicate model errors.
```

```python
linear_assumption(model3,X_test3,y_test3)
```

```python
def homoscedasticity_assumption(model, features, label):

    df_results=calculate_residuals(model,features,label)
    #plotting the residuals
    plt.subplots(figsize=(10,8))
    ax=plt.subplot(111) #to remove spines
    plt.scatter(x=df_results.index,y=df_results.residuals,alpha=0.5,color="green") 
    plt.plot(np.repeat(0,df_results.index.max()),color="brown",linestyle="--")
    plt.gca().set_facecolor('#f5f5dc')
    plt.title("Residual Plot")
    plt.xlabel("Index")
    plt.ylabel("Residuals")
    
    plt.show()

# Checks homoscedasticity by plotting residuals (errors) against index values.
# Random scatter around zero suggests constant variance, supporting homoscedasticity.
# Patterns or funnel shapes indicate heteroscedasticity, meaning variance is not constant.
```

```python
homoscedasticity_assumption(model3,X_test3,y_test3)
```
