```python
# task1:
# 1.only Load the dataset and display it in 2 lines code
import pandas as pd
df = pd.read_csv('DS1_C4_S8_Tendulkar_ODI_Data_Practice.csv')
df
```

```python
# 2. Extract the following columns of the dataset to perform the analysis: a)Runs b)BF c)4s d)6s
df = df[['Runs','BF','4s','6s']]
df
```

```python
# 3. Drop the null values of these extracted columns:Runs, BF, 4s, 6s.
df = df.dropna(subset=['Runs','BF','4s','6s'])
df
```

```python
# 4. Check data for the extracted columns and convert them into the correct data types for further analysis.
df.info()
df['Runs'] = df['Runs'].astype(int)
df['BF'] = df['BF'].astype(int)
df['4s'] = df['4s'].astype(int)
df['6s'] = df['6s'].astype(int)
df.info()
df.head(2)
```

```python
# task2
# After importing the required libraries, calculate mean, median, and mode scores for the extracted columns: Runs, BF, 4s, and 6s.
import numpy as np
mean = df[['Runs', 'BF', '4s', '6s']].mean()
median = df[['Runs', 'BF', '4s', '6s']].median()
mode = df[['Runs', 'BF', '4s', '6s']].mode()
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
# task3
# Calculate the variance and standard deviation for the extracted columns: Runs, BF, 4s, and 6s.
variance = df[['Runs', 'BF', '4s', '6s']].var()
std_dev = df[['Runs', 'BF', '4s', '6s']].std()
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
# task4
# Calculate kurtosis for the extracted columns: Runs, BF, 4s, and 6s.
from scipy.stats import kurtosis
kurtosis = df[['Runs', 'BF', '4s', '6s']].apply(kurtosis)
print("Kurtosis:", kurtosis)
```

```python
# task5
# Use box plots to present the distribution for the extracted columns: Runs, BF, 4s, and 6s.
import matplotlib.pyplot as plt
df[['Runs', 'BF', '4s', '6s']].plot(kind='box')
plt.show()
```
