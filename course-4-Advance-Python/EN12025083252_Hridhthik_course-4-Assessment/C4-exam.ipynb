```python
# 2. Import the Pandas library and create a Series named "temperatures" with the 
# following data and custom index: [10, 15, 20, 25], ["Monday", "Tuesday", 
# "Wednesday", "Thursday"] 
index=["Monday", "Tuesday","Wednesday", "Thursday"]
data=[10, 15, 20, 25]
temperature=pd.Series(data,index=index)
temperature
```

```python
# 3. Calculate the mean, standard deviation, and size of the temperatures Series which is 
# created in question 2. 
# Mean
Mean=temperature.mean()
print('Mean:',Mean)
Std=temperature.std()
print('Standard deviation:',Std)
print('size:',temperature.size)
```

```python
# 6. Perform inner join on the above two dataframes X and Y
data_X = {'Roll No': [10, 20, 30, 40], 'name': ['a', 'b', 'c', 'd']}
X = pd.DataFrame(data_X)
print(X)
data_Y = {'Roll No': [10, 20, 90, 80], 'name': ['p', 'q', 'r', 's']}
Y = pd.DataFrame(data_Y)
print(Y)
```

```python
merged_df = pd.merge(X, Y, on="Roll No", how="inner")
merged_df
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Example data for line and bar charts
line = pd.DataFrame({
    'X': [1, 2, 3],
    'Y': [10,15,12]
})

bar = pd.DataFrame({
    'Categories': ['A', 'B', 'C'],
    'Values': [25,30,20]
})

# Create subplots
fig, ax = plt.subplots(1, 2, figsize=(15, 7))

# Line chart on the first subplot
ax[0].set_title('Line Chart')
ax[0].plot(line['X'], line['Y'])

# Bar chart on the second subplot
ax[1].set_title('Bar Chart')
ax[1].bar(bar['Categories'], bar['Values'])

# Display the plot
plt.show()
```
