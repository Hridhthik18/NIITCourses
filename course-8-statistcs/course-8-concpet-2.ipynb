```python
import statistics as st
import scipy.stats as sts
```

```python
data_d=emp_df['Annual_Salary($)'].tolist()
```

```python
p_mean=st.mean(data_d)
p_median=st.median(data_d)
p_std=st.stdev(data_d)
```

```python
skewness=sts.skew(data_d,bias=False)
```

```python
kurtosis=sts.kurtosis(data_d,bias=False)
```

```python
print('mean',p_mean)
print('median',p_median)
print('standard_deviation',p_std)
print('skewness',skewness)
print('kurtosis',kurtosis)
```

```python
sal_mean=st.mean(sal_40)
sal_median=st.median(sal_40)
sal_std=st.stdev(sal_40)
skewness_sal=sts.skew(sal_40,bias=False)
kurtosis_sal=sts.kurtosis(sal_40,bias=False)


print('mean',sal_mean)
print('median',sal_median)
print('standard_deviation',sal_std)
print('skewness',skewness_sal)
print('kurtosis',kurtosis_sal)


```
