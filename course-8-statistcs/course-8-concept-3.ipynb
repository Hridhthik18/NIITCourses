```python
import pandas as pd
import statistics as st
import matplotlib.pyplot as plt
import scipy.stats as sc
import math 
```

```python
# step 1 :establish  the null and alternative hypothesis
H0: average size of us  farm  is less than or equal to 445.5 acres in 2019
H1:average size of us  farm  is more than 445.5 acres in 2019
```

```python
# step2:symbolize
H0:p<=445.4
H1:p>445.5
# so here we have to perform right tailed test
```

```python
# population mean 445.5
```

```python
# step3:reading the file
df=pd.read_csv('DS1_C8_S3_FarmSize_Data_Concept.csv')
df
```

```python
# information about data set
sample size n=35
population mean=445.5
population standard deviation=466.827
alpha =5%=0.05
```

```python
#step4: establish the decision rule
if pvalue<alpha :reject the  null hypothesis
```

```python
#step5: gather the sample data of 35 size
sample_df=df.sample(n=35,random_state=1)
sample_df
```

```python
# step6:calculate the sample mean
s_mean=st.mean(sample_df['2019_Average_farm_size(acres)'])
print('sample_mean=',s_mean)
```

```python
# step7:analyze data
from scipy.stats import norm
x=421.4 #sample_mean
mu=445.5 #population mean
S=466.827 #std 
n=35 #sample size
alpha=0.05 #significance level
```

```python
z_statistic = (x-mu) / (S/math.sqrt(n)) 
print("The Z statistics is ", z_statistic)

p_value = norm.sf(abs(z_statistic))       #sf is a survival function -complementaru unfiform distribution function
print("The p_value is "+str(p_value))     #p value reprsents the probabilty of observing 

z_critical = norm.ppf(1-alpha)# Right Tailed test   inverse of the cummulative distribution function
print("The z-critical value is "+str(z_critical))
```

```python
Decision Rule: Reject the Null hypothesis

If the calculated Z-statistic is greater than the Z-critical value or if the p-value is less than the significance level (alpha), you can reject the null hypothesis.

1.  Z-statistic > Z-critical or
2.   pvalue < alpha

So, from the above analysis:
p-value = 0.38 and alpha = 0.05 : pvalue> alpha : Accept the Null Hypothesis
z-statistics = -0.3054 , Z-critical = 1.644 : Z-statitics<Z-critical : Accept the Null Hypothesis

Buisness Decision                                                                   
So , we can conclude that Average size of a U.S farm is less than or equal to 445.5 acres in 2019. There is no significance change in the farms size within a year and hypothesis testing is also identified the same.
```
