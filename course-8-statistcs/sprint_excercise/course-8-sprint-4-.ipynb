```python
import pandas as pd
import statistics as st
from scipy.stats import t 
import math
```

```python
# task1
# a
H0: The mean score of the class test is less than or equal to 71.
H1:The mean score of the class test is greater than 71.

    
H0:p<=71
H1:p>71

This is a one-tailed hypothesis test as we are testing if the mean score is greater than 71.
```

```python
# b
sample_data=[61,93,75,67,84,95]

s_mean=st.mean(sample_data)
s_std=st.stdev(sample_data)
s_var=s_std**2
n=len(sample_data)
df=n-1
alpha=0.1

print('sample_mean:',s_mean)
print('sample_std:',s_std)
print('sample_var:',s_var)
print('length of sample:',n)

t_critical=t.ppf(1-alpha,df)
print('t_critcal_value:',t_critical)
```

```python
# c
p_mean=71

t_statistics=(s_mean-p_mean)/(s_std/math.sqrt(n))
print('t_statistics:',t_statistics)

p_value=t.sf(abs(t_statistics),df)
print('p_value:',p_value)
```

```python
# d
import matplotlib.pyplot as plt
import numpy as np

s_mean=79.166
s_std=13.862
upper=s_mean+4*s_std
lower=s_mean-4*s_std
x=np.arange(-4,4,0.01)
tvalues5=t.pdf(x,5)
plt.plot(tvalues5)
plt.legend(['df=5'])
plt.show()
```

# conclusion

alpha=0.1), we fail to reject the null hypothesis.

This means there is insufficient evidence to conclude that the mean score for the class is greater than 71 at a 90% confidence level.
