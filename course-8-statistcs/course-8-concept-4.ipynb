## Demo – 4.1: Hypothesis Test Using Single Sample Mean

The annual per-patient average expenditure on dental care is around $1,272 in a certain country. The regional dental association wants to test to determine if this figure is accurate for their area of the country. To test this, 12 families of four members are randomly selected from the population in that area of the country and a log is kept of the family’s dental expenditures for one year. The collected sample data are as follows. Assume the dental expenditures are normally distributed in the population, use a 5% significance level to test the dental association’s hypothesis.

1208,812,1217,1323,1308,1515,831,1021,1387,951,930,730

### Establish the null and alternate hypothesis

```python
H0: the average dental care expenditure of a family of four is $1272 in the country.
H1: the average dental care expenditure of a family of four is more or less than $1272 in the country.
H0:p=1272
H1:p!=1272
```

### Calculate Sample Mean and standard deviation

```python
import statistics as st
sample_data=[1208,812,1217,1323,1308,1515,831,1021,1387,951,930,730]
s_mean=st.mean(sample_data)
s_std=st.stdev(sample_data)
n=len(sample_data)

print('sample_mean:',s_mean)
print('sample_std:',s_std)
print('sample_size',n)
```

### Determine the appropriate statistical test

```python
p_mean=1272
s_mean=1102.75
s_std=256.26624117046
n=12
df=n-1
```

### Set the value of alpha

```python
alpha=0.05
```

### Establish the decision rule

```python
if pvalue < alpha :reject the null hypothesis
if -t-critical>t-statistcs>t-critical :reject null hypothesis
```

### Analyze the data

```python
import math
from scipy.stats import t

t_statistic = (s_mean - p_mean) / (s_std/math.sqrt(n)) 
print("The t-statistic is ", t_statistic)

p_value=t.sf(abs(t_statistic),df)*2 # considering two tails
print("The p_value is ", p_value)

t_critical = t.ppf(alpha/2, df)
print("The t-critical value is "+str(t_critical))
```

### Reach a statistical conclusion

```python
p_value=0.0429 < alpha =0.05 :which leads to the rejection of null hypothesis
t_statistic=-2.287 < t_critical=-2.20 which lead to  the rejection of null hypothesis
```

## Demo – 4.2: Hypothesis Test Using Two Sample Means

Consider a survey was conducted by randomly sampling self-initiated expatriates in two industries: information technology (IT) and finance. Each one is asked to rate his or her present job satisfaction on a 7-point Likert scale, with 7 being very satisfied and 1 being very unsatisfied.
If we need to find whether there is a difference in the above job satisfaction ratings. Using a 5% significance level, determine whether there is a significant difference in the mean job satisfaction ratings of the two industries. 
Assume that the ratings for this test are normally distributed and that the population variances are approximately equal.

### Establish the null and alternate hypothesis

```python
HO: the difference between two sample means is zero.
H1: the difference between the two-sample means is not equal to zero.

H0:p1=p1
H1:p1!=p2
```

### Determine the appropriate statistical test

We will test hypothesis about a population mean using the t-statistics (standard deviation of population is known).
t-statistic = [(𝑥1- 𝑥2) – (μ1 - μ2)] / [sqrt((s1^2(n1 -1) + s2^2(n2-1))/df)*sqrt(1/n1 + 1/n2)] Where, df is the degrees of freedom.

### Set the value of alpha (α)

```python
alpha=0.05
```

### Establish the decision rule

### Set degrees of freedom

```python
sample_IT=[4,5,6,4,4,3,2,5,3,4,7]
sample_Finance=[2,3,4,5,4,7,6,4,3]

x1=st.mean(sample_IT)
x2=st.mean(sample_Finance)
v1=st.stdev(sample_IT)
v2=st.stdev(sample_Finance)
n1=len(sample_IT)
n2=len(sample_Finance)
df=n1+n2-2
print(df)
```

### Analyze the data

#### Calculate sample statistic

```python
sample_IT=[4,5,6,4,4,3,2,5,3,4,7]
sample_Finance=[2,3,4,5,4,7,6,4,3]

x1=st.mean(sample_IT)
x2=st.mean(sample_Finance)
s1=st.stdev(sample_IT)
s2=st.stdev(sample_Finance)
n1=len(sample_IT)
n2=len(sample_Finance)
v1=s1**2
v2=s2**2


print('sample_mean for it:',x1)
print('sample mean for finace',x2)
print('std_IT:',s1)
print('std_FI:',s2)
print('Variance_IT:',v1)
print('Variance_Finance:',v2)
print('length_IT:',n1)
print('length_Finance',n2)
```

#### Calculate t-statistic and t-critical

```python
noise = math.sqrt(v1*(n1-1) + v2*(n2-1))*math.sqrt(1/n1 + 1/n2)
signal = abs(x1-x2)
t_statistic=signal/noise
print('t-statistic =',t_statistic)

p_value=t.sf(abs(t_statistic),df)*2
print("The p_value is ", p_value)

t_critical = t.ppf(1-0.025, df)
print('t-critical =',t_critical)
```

### Reach a statistical conclusion

```python
P_value=0.9859 >alpha =0.05 which leads to acceptance of null hypothesis
```
